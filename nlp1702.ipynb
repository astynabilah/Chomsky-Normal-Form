```
import nltk
from nltk import CFG
from nltk.parse.generate import generate

grammar1 = CFG.fromstring("""
  S -> NP VP
  VP -> V NP | V NP PP
  PP -> P NP
  V -> "melihat" | "makan" | "berjalan"
  NP -> "Cika" | "Ciko" | "Cino" | Det N | Det N PP
  Det -> "sebuah" | "seorang"
  N -> "orang" | "anjing" | "kucing" | "teleskop" | "taman"
  P -> "di" | "dengan" | "bersama" 
""")
# check if in Chomsky Normal Form
CFG.is_chomsky_normal_form(grammar1)

sentence = ['Cino', 'melihat', 'seorang' , 'Ciko'] 
parser = nltk.ChartParser(grammar1)
for tree in parser.parse(sentence):
  print(tree)

for sentence in generate(grammar1, n=10):
  print(' '.join(sentence))
  # parser = nltk.ChartParser(grammar1)
  # for tree in parser.parse(sentence):
  #   print(tree)

```
